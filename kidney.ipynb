```python
df.drop(['id'],axis=1,inplace=True)
```

```python

df.drop(['age','bp','su', 'rbc', 'pc', 'pcc', 'ba', 'bgr','bu','sod', 'pot','wc', 'rc','dm', 'cad','appet', 'pe', 'ane'],axis=1,inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
s=sns.pairplot(df)
```

```python
s=sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

```python
s=sns.heatmap(df.corr(),annot=True)
```

```python
x,y=df.drop('classification',axis=1), df['classification']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)*100
```

```python
import tkinter as tkn
app=tkn.Tk()
app.geometry('400x400')
app.title('Chronic Kidney Disease Predictor')
mod= pd.read_pickle('Chronic_Kidney_disease_predictor.pkl')
             
sg = tkn.Variable(app)
al = tkn.Variable(app)
sc = tkn.Variable(app)
hemo= tkn.Variable(app)
pcv= tkn.Variable(app)
htn= tkn.Variable(app)
result=tkn.Variable(app)
             
tkn.Label(app,text='Specific Gravity:    ',font=('Arial',15)).place(x=10,y=20)
tkn.Label(app,text='Albumin:    ',font=('Arial',15)).place(x=10,y=50)
tkn.Label(app,text='Serum Creatinine:    ',font=('Arial',15)).place(x=10,y=80)
tkn.Label(app,text='Hemoglobin:    ',font=('Arial',15)).place(x=10,y=110)
tkn.Label(app,text='Packed Cell Volume:    ',font=('Arial',15)).place(x=10,y=140)
tkn.Label(app,text='Hypertension:    ',font=('Arial',15)).place(x=10,y=170)
tkn.Label(app,textvariable=result,font=('Arial',15),fg='red').place(x=150,y=200)

tkn.Entry(app,textvariable=sg,width=12,font=('Arial',15)).place(x=200,y=20)
tkn.Entry(app,textvariable=al,width=12,font=('Arial',15)).place(x=200,y=50)
tkn.Entry(app,textvariable=sc,width=12,font=('Arial',15)).place(x=200,y=80)
tkn.Entry(app,textvariable=hemo,width=12,font=('Arial',15)).place(x=200,y=110)
tkn.Entry(app,textvariable=pcv,width=12,font=('Arial',15)).place(x=200,y=140)
tkn.Entry(app,textvariable=htn,width=12,font=('Arial',15)).place(x=200,y=170)

def prediction():
    global mod
    spec=eval(sg.get())
    alb=eval(al.get())
    seri=eval(sc.get())
    hem=eval(hemo.get())
    pack=eval(pcv.get())
    hyp=eval(htn.get())
    query=[[spec,alb,seri,hem,pack,hyp]]
    result.set(round(mod.predict(query)[0],0))
    
    
tkn.Button(app,text='Predict',font=('Arial',15),bg='white',command=prediction).place(x=150,y=240)

app.mainloop()
```
